```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(5) # creates an array with elements ranging from 0 to 4
print(x)
```

```python
y = x
print(y)
```

```python
plt.plot(x, y, 'o')
plt.show()
plt.plot(x, y, 'o-')
plt.show()
plt.plot(x, y, 'o--')
plt.show()
```

```python
plt.plot(x, y, 'o--')
plt.plot(x, -y, 'o-')
plt.title('y = x and y = -x')
plt.show()
```

```python
N = 10
x = np.linspace(0, 10, N) # It automatically calculates the step size based on N
print(x)
y = x
plt.plot(x, y, 'o--')
plt.axis('off') # only visualization but no legend
plt.show()
```

```python
y = np.logspace(0.1, 1, N) # y changes in logarithmic order
print(y)
plt.plot(x, y, 'o--')
plt.show()
```

```python
y = np.geomspace(0.1, 1000, N) # geometric space
print(y)
plt.plot(x, y, 'o--')
plt.show()
```
